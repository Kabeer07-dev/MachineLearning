```python
numeric_feature = ['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']

def plotting(var,num):
    plt.figure(figsize=(12,8))
    plt.subplot(2,3,num)
    sns.histplot(df[var],kde=True)


for i,feature in enumerate(numeric_feature):
    plotting(feature,i+1)
```
