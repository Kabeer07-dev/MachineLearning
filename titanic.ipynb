### Data Visualization

```python
def plotting(var,num):
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,num)
    sns.histplot(df[var],kde=True)


numeric = ['Age','Fare']

for i,var in enumerate(numeric):
    plotting(var,i+1)

```

```python
df.columns
```

```python
categorial = ['Pclass','Sex','SibSp','Parch','Embarked']

def count_plot(var,num):
    plt.figure(figsize=(8,12))
    plt.subplot(3,2,num)
    sns.countplot(x=df[var])


for i,var in enumerate(categorial):
        count_plot(var,i+1)
```

```python
gender_Survived = df.groupby('Sex')['Survived'].mean() * 100
gender_Survived
```

```python
class_survived = df.groupby('Pclass')['Survived'].mean()*100
class_survived
```

```python
gender_Survived.plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate (%)")
plt.xticks(rotation=0)
plt.ylim(0, 100)

plt.show()
```

```python
class_survived.plot(kind='bar')
plt.title('Survival rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate (%)')
plt.xticks(rotation=0)
plt.ylim(0,100)

plt.show()
```

```python
q1 =  df2['Fare'].quantile(0.25)
q3 =  df2['Fare'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
lower_bound = q1 - 1.5*(iqr)
upper_bound = q3 + 1.5*(iqr)

outliers = df2[(df2['Fare'] < lower_bound) | (df2['Fare'] > upper_bound)]
print(len(outliers))
```

### Encoding attributes

```python
df2['Sex'] = df2['Sex'].map({
    'male':0,
    'female':1
})

df2.head()
```

```python
df2 = df2.rename(columns={
    'Sex':'is_female'
})

df2.head()
```

```python
df2 = pd.get_dummies(df2,columns=['Embarked'],drop_first=True)
df2.head()
```

### scaling

```python
df2[['Embarked_Q','Embarked_S']] = df2[['Embarked_Q','Embarked_S']].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaling_attr = ['Pclass','Age','Fare']

scalar = StandardScaler()

df2[scaling_attr] = scalar.fit_transform(df2[scaling_attr])
```
